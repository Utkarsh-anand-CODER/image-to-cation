```python
caption_path='C:/Users/KIIT/OneDrive/Desktop/minor_project/archive (5)/Flickr_Data/Flickr_TextData/Flickr8k.token.txt'
```

```python
captions=open(caption_path, 'rb').read().decode('utf-8').split('\n')[:-1]
```

```python
captions
```

```python
captions[-1]
```

```python
first,second=captions[0].split('\t')
print(first)
```

```python
captions_dict = {}
for i in captions:
    first,second=i.split('\t')
    img_name=first.split(".")[0]
    if captions_dict.get(img_name) is None:
        captions_dict[img_name]=[]
    captions_dict[img_name].append(second)
    

        
```

```python
captions_dict
```

```python
def preprocessed(txt):
    modified=txt.lower()
    modified='startofseq' + modified + 'endofseq'
    return modified
```

```python
for k,v in captions_dict.items():
    for vv in v:
        captions_dict[k][v.index(vv)]=preprocessed(vv)
```

```python
count_words={}
count=1
for k,vv in captions_dict.items():
    for v in vv:
        for word in v.split():
            if word not in count_words:
                count_words[word]=count
                count += 1
```

```python
count_words
```

```python
for k, vv in captions_dict.items():
    for v in vv:
        encoded = []
        for word in v.split():  
            encoded.append(count_words[word])
        captions_dict[k][vv.index(v)] = encoded

        
```

```python
captions_dict
```

```python
MAX_LEN=0
for k,vv in captions_dict.items():
    for v in vv:
        if len(v)>MAX_LEN:
            MAX_LEN=len(v)
            print(v)
```

```python
from keras.utils import to_categorical
from keras_preprocessing.sequence import pad_sequences
```

```python
VOCAB_SIZE=len(count_words)

def generator(photo,caption):

    
    x=[]
    y_in=[]
    y_out=[]
    
    for k,vv in caption.items():
        for v in vv:
            for i in range(0,len(v)):
                x.append(photo[k])
                
                in_seq=[v[:i]]
                out_seq=v[i]
                
                in_seq = pad_sequences(in_seq,maxlen=MAX_LEN,padding='post',truncating='post')[0]
                out_seq=to_categorical([out_seq], num_classes=VOCAB_SIZE+1)[0]
                
                y_in.append(in_seq)
                y_out.append(out_seq)
    return x, y_in, y_out
```

```python
x, y_in, y_out = generator(images_features, captions_dict)
```
